```python
#PartA
import numpy as np

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV

from keras.layers import Dense, Activation
from keras.models import Sequential
from keras.utils import np_utils
```

```python
#PartB
from keras.datasets import mnist
(X_train, y_train), (X_test, Y_test) = mnist.load_data()
X_train = X_train.reshape(X_train.shape[0], 784)
X_test = X_test.reshape(X_test.shape[0], 784)
print(X_train.shape)
print (X_test.shape)
```

```python
y_train
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255

```

```python
#PartC
y_train = np_utils.to_categorical(y_train, 10)
Y_test = np_utils.to_categorical(Y_test, 10)
```

```python
print (y_train.shape)
print (y_train[:10,:])
```

```python
#PartD
input_size = 784
hidden_neurons = 100
out_size = 10
def model_creator():
    model = Sequential()
    model.add(Dense(hidden_neurons, input_dim=input_size))
    model.add(Activation('sigmoid'))
    
    model.add(Dense(out_size, input_dim=hidden_neurons))
    model.add(Activation('softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics = ['accuracy'])
    return model
```

```python
#PartE
seed = 2
np.random.seed(2)
```

```python
#PartF
model = KerasClassifier(build_fn=model_creator, verbose = 2)

```

```python
#PartG
epochs = np.array([10, 15, 20])
batch_size = np.array([30, 50, 100])
param_grid = dict(epochs = epochs, batch_size = batch_size)
grid = GridSearchCV(estimator=model, param_grid=param_grid, scoring='neg_log_loss', cv=10)
```

```python
grid_result = grid.fit(X_train, y_train, verbose = 1)
```

```python
#PartH
#run this after getting all grid results.
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
score = grid.best_estimator_.model.evaluate(X_test, Y_test)
print("Score: " + str(score[1]))
```
